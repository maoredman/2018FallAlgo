```python
infile = open('pg2_resource/5000.in')
n = int(infile.readline().strip()) // 2
point2point = {}
for idx, line in enumerate(infile):
    if idx == n:
        break
    chord = [int(i) for i in line.strip().split(' ')]
    point2point[chord[1]] = chord[0]
    point2point[chord[0]] = chord[1]
# print(point2point)
```

```python
saved_mps = {}
saved_mps_chords = {}
for idx in range(2*n): # initialize
    saved_mps[(idx, idx)] = 0
    saved_mps[(idx, idx-1)] = 0
    saved_mps_chords[(idx, idx)] = []
    saved_mps_chords[(idx, idx-1)] = []

for j in range(1, 2*n):
    k = point2point[j]
    for i in range(j):
        if i <= k and k <= j-1 and saved_mps[(i,k-1)]+1+saved_mps[(k+1,j-1)] > saved_mps[(i,j-1)]:
            saved_mps[(i,j)] = saved_mps[(i,k-1)]+1+saved_mps[(k+1,j-1)]
            saved_mps_chords[(i,j)] = [(i,k-1), (k,j), (k+1, j-1)]
        else:
            saved_mps[(i,j)] = saved_mps[(i,j-1)]
            saved_mps_chords[(i,j)] = [(i, j-1)]

print('max:', saved_mps[(0,2*n-1)])
def print_chords(i,j):
    if len(saved_mps_chords[(i,j)]) == 3:
        print_chords(saved_mps_chords[(i,j)][0][0],saved_mps_chords[(i,j)][0][1])
        print(saved_mps_chords[(i,j)][1])
        print_chords(saved_mps_chords[(i,j)][2][0],saved_mps_chords[(i,j)][2][1])
    elif len(saved_mps_chords[(i,j)]) == 1:
        print_chords(saved_mps_chords[(i,j)][0][0],saved_mps_chords[(i,j)][0][1])
    else:
        assert(len(saved_mps_chords[(i,j)]) == 0)
print_chords(0,2*n-1)
```
